```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 

june_temps = []
june_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6)
```

```python
#  3. Convert the June temperatures to a list.
june_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
june_temps
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps, columns=['date','June tobs'])
june_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.

dec_temps = []
dec_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)== 12)
```

```python
# 7. Convert the December temperatures to a list.
dec_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)== 12).all()


dec_temps
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_df = pd.DataFrame(dec_temps, columns=['date','Dec tobs'])
dec_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_df.describe()
```

```python
# Additional query: Query for precipitation for the month of june
june_prcp = []
june_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract("month", Measurement.date)==6).all()
june_prcp
```

```python
# Create a DataFrame for the list of recorded 'prcp' for June and provide summary statistic.
june_rain_df = pd.DataFrame(june_prcp, columns=['Date','June prcp'])
june_rain_df.describe()
```

```python
# Additional query: Query for precipitation for the month of Dec
dec_prcp = []
dec_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract("month", Measurement.date)==12).all()
dec_prcp
```

```python
# Create a DataFrame for the list of recorded 'prcp' for Dec and provide summary statistic.
dec_rain_df = pd.DataFrame(dec_prcp, columns=['Date','Dec prcp'])
dec_rain_df.describe()
```

```python
# Graph June precipitation records

june_rain_df.plot.hist(bins=20)
plt.tight_layout()
```

```python
# Graph Dec precipitation records
dec_rain_df.plot.hist(bins=20)

plt.tight_layout()
```

```python
# Create a summary dataframe of june vs Dec precipitation

prcp_summary = june_rain_df.merge(dec_rain_df, left_index=True, right_index=True)
prcp_summary
```

```python
# Create a summary dataframe of june vs Dec temps
june_tobs = june_df.describe()
dec_tobs = dec_df.describe()

temp_summary = june_tobs.merge(dec_tobs, left_index=True, right_index=True)
temp_summary
```
